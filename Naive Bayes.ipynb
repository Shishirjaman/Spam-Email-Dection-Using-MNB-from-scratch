```python
import math
import pandas as pd

totalmail = 0
totalham = 0
totalspam = 0

mailcount = {}
types_of_words_cnt= {}
prob_of_ham_spam= {}
likelyhood= {} 

vocabulary = set()


print("data initializing.......(training_data)\n\n")

#train_data= pd.read_csv('train.csv')    "I try to use pandas module to read file but I always get a keyword error while i try to add number of times a word exist in words types."
train_data =open('train.csv','r')

for i in train_data:
    words = i.split(" ")[2:]   
    for (index, word) in enumerate(words):
        if index % 2 == 0:
            words += [word]

    for item in words:
        vocabulary.add(item)

train_data.close()
print("data initialized.........\n\n")
    
print("Getting Probablity of ham and spam from training data........\n\n")

types_of_words_cnt.setdefault("spam", {})
types_of_words_cnt.setdefault("ham", {})

for word in vocabulary:
    types_of_words_cnt["spam"].setdefault(word, 0)
    types_of_words_cnt["ham"].setdefault(word, 0)

#train_data = pd.read_csv('train.csv')    
train_data =open('train.csv','r')

for i in train_data:
    totalmail += 1
    lines = i.split(" ")
    Type = lines[1]
    mailcount.setdefault(Type, 0)
    mailcount[Type] += 1

    for j in range(2, len(lines), 2):
        types_of_words_cnt[Type][lines[j]] += int(lines[j+1]) # at this line i get keyword eror if i use pandas module

    
for word, count in types_of_words_cnt["spam"].items():
    totalspam += count

for word, count in types_of_words_cnt["ham"].items():
     totalham += count

for Type, count in mailcount.items():
    prob_of_ham_spam.setdefault(Type, 0.0)
    prob_of_ham_spam[Type] = float(count/totalmail)
    print("probablity: "+str(float(count/totalmail)))

print("\nGetting likelyhood ..............\n\n")
likelyhood.setdefault("spam", {})
likelyhood.setdefault("ham", {})

for Type, attribute in types_of_words_cnt.items():
    for word, count in attribute.items():
        likelyhood[Type].setdefault(word, 0)
        likelyhood[Type][word] = float((count + 1) / float(mailcount[Type] + len(vocabulary)))
            
train_data.close()



print("initializing and predicting mail type of the test data......\n\n")

#test_data = pd.read_csv('test.csv')
test_data =open('test.csv','r')
output_data = open('output.csv','w')

actualresult = list()
calculatedresult = list()

print("****Predicted output****\n")
print("Mail_ID   Predicted_Type\n\n") 

for i in test_data:
    words = i.split(" ")
    mail_id = words[0]
    actual = words[1]

    predicted = ""
    wordList = []
    for index, word in enumerate(words[2:]):
        if index % 2 == 0:
            wordList += [word]

    Ham = math.log10(prob_of_ham_spam["ham"])
    Spam = math.log10(prob_of_ham_spam["spam"])
    
       
    for word in wordList:
        if word not in vocabulary:  
            Ham += (1 / float(math.log10(totalham))) *(1/float(math.log10(len(vocabulary)))) # takes a tends to zero value in lieu of 
            Spam += (1 / float(math.log10(totalspam))) *(1/float(math.log10(len(vocabulary)))) # zero to increase accuracy.
        else:
            Ham += math.log10(likelyhood["ham"][word])   # use log to reduce overflow in lieu of Ham= Ham*likelyhood
            Spam += math.log10(likelyhood["spam"][word])

    if Spam > Ham:
            predicted = "spam"
    else:
            predicted = "ham"

    actualresult.append(actual)
    calculatedresult.append(predicted)
    
    output_data.write(mail_id + " " + predicted + "\n")
    print(f"{mail_id}  {predicted}\n")
    
test_data.close()
output_data.close()



print("\ntesting accuracy......\n")

crrSpam = 0
crrHam = 0
incrrSpam = 0
incrrHam = 0


for actual,predicted in zip(actualresult,calculatedresult):
    if (predicted in "spam"):
        if(actual in "spam"):
            crrSpam += 1
        elif(actual in "ham"):
            incrrSpam += 1

    elif (predicted in "ham"):
        if(actual in "ham"):
            crrHam += 1
        elif (actual in "spam"):
            incrrHam += 1

precision = float(crrSpam/float(crrSpam+incrrSpam)) * 100
recall = float(crrSpam/float(crrSpam+incrrHam)) * 100
f1_score = (2*precision*recall)/(precision+recall) 

print("\nPrecision: " + str(precision)+"\n")
print("Recall: " + str(recall)+"\n")
print("F1 Score: "+ str(f1_score)+"\n")

```
